```python
ID = [1,2,3,4,5,6,7,8,9]
sales = [11,22,33,44,55,66,77,88,99]
profit = [15,25,35,45,55,65,75,85,95]
revenue = [17,37,27,77,49,58,98,45,115]

```

LINE CHART

```python
plt.plot(ID,sales, color="r",label="sales",marker="o")

plt.plot(ID,profit, color="y",label="profit",marker="o")

plt.plot(ID,revenue, color="b",label="revenue",marker="o")

plt.xlabel("ID")
plt.ylabel("metrics")
plt.title("Financial Report")

# plt.legend(["SALES","PROFIT"])
plt.legend()
plt.grid(True)

plt.savefig("financial report.png") 

plt.show()
```

BAR CHART

```python
plt.bar(ID,sales, color="#370726",label="sales")

plt.plot(ID,profit,color="r", label="profit",marker="o")

plt.plot(ID,revenue, color="g", label="revenue",marker="o")


plt.xlabel("ID")
plt.ylabel("metrics")
plt.title("Financial Report")

plt.legend()
plt.grid(True)
plt.show()
```

```python
ID = [11,22,33,44,55,66,77,88,99]
sales = [11345,22789,35673,46704,55890,66452,77167,89808,91239]
profit = [13455,20865,32535,47835,55630,65123,75730,83985,92625]
revenue = [17678,37235,27907,76807,47209,53638,92768,45463,13615]

x_indexes=np.arange(len(ID))
width=0.25

plt.bar(x_indexes-width,sales,width=width, color="#370726",label="sales")

plt.bar(x_indexes,profit, width=width,color="r", label="profit")

plt.bar(x_indexes+width,revenue,width=width, color="g", label="revenue")


plt.xlabel("ID")
plt.ylabel("metrics")
plt.title("Financial Report")

plt.xticks(ticks=x_indexes,labels=ID)

plt.legend()
plt.grid(True)
plt.show()
```

PIE CHART

```python
ages = [10,20,30,40,50]
name = ["chiamaka","gbemi","lubabat","nuh","basirah"]

explode= [0.3,0,0,0,0]
colors = ["gold","brown","red","purple","blue"]

plt.pie(ages, labels= name, colors=colors, 
        explode=explode,shadow=True, wedgeprops={"edgecolor":"black"},startangle=90, autopct="%1.1f%%")
```

HISTOGRAM

```python
ages = [18,19,21,25,26,30,32,38,45,55]
bins = [10,20,30,40,50,60]



plt.hist(ages, bins=bins, color="green", edgecolor="black")

plt.title("Respondents")
plt.xlabel("Ages")
plt.ylabel("total number of respondents (frequency)")
# plt.grid(True)
```
