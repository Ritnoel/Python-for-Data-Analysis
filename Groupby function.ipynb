```python
import pandas as pd
import numpy as np
```

```python
data = {
    "year": [2010,2010,2010,2010,2011,2011,2011,2011],
    "company":["coca-cola","dangote","unilever","nestle","coca-cola","dangote","unilever","nestle"],
    "sales_person":["amaka","rit","comfort","wura","abiola","aliyu","ikrama","mercy"],
    "sales":[200,300,400,500,600,700,800,900]
}
df=pd.DataFrame(data)
df
```

```python
df.groupby("company").sum()
```

```python
total_sales=df.groupby("year").sum()
total_sales
```

```python
##exporting as a csv file
df.to_csv("total_sales.csv",index=False)
```

```python
##exporting as an excel file
df.to_excel("total_sales.xlsx")
```
