```python
##join
##left join
##right join
##inner join
##full or outer join
```

```python
Employees = pd.DataFrame({
    "ID": [1,2,3,4,5,6],
    "Name": ["chiamaka", "gbemi", "mojisola","wuraola","aliyu","mercy"],
    "State":["imo", "kogi", "osun", "ondo", "kano", "lagos"],
    "Salary":[2300,5000,7800,2000,6700,9000]
})

Department = pd.DataFrame({
    "ID":[2,5,4,1,6],
    "dept":["sales","hr","marketing","engineering","finance"]
})
```

```python
Employees
```

```python
Department
```

```python
##inner join
```

```python
innerjoin_df = pd.merge(Employees,Department, on="ID",how="inner")
innerjoin_df
```

```python
##outer join
```

```python
outerjoin_df = pd.merge(Employees, Department, on="ID", how="outer")
outerjoin_df
```

```python
##left join
```

```python
leftjoin_df = pd.merge(Employees, Department, on="ID", how="left")
leftjoin_df
```

```python
leftjoin_df = pd.merge(Department,Employees, on="ID", how="left")
leftjoin_df
```

```python
rightjoin_df = pd.merge(Employees, Department, on="ID", how="right")
rightjoin_df
```

```python
##concatenation
```

```python
df = pd.concat([Employees,Department])
df
```

```python
df = pd.concat([Employees,Department], axis=1)
df
```

```python
df = pd.concat([Employees,Department], axis=1,join="inner")
df
```
