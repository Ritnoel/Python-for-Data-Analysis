```python
import pandas as pd
import numpy as np
```

```python
data = {
    "names":["esther", "rit", "chiamaka", "lubabat", "aliyu","esther"],
    "marks":[30,40,60,55,90,30],
    "city":["lagos","kano","kogi","kaduna","plateau","lagos"]
}
```

```python
df = pd.DataFrame(data, index=["A","B","C","D","E","F"])
df
```

```python
df.drop_duplicates()
```

```python
df
```

```python
df.drop_duplicates(subset="names", keep="last", inplace=True)
```

```python
df
```

##sorting

```python
##sorting a single column
sort = df.sort_values(by="names", ascending=False)
sort
```

```python
##sorting multiple columns
sort2 = df.sort_values(by=["names","city"], ascending=False)
sort2
```

```python
##sorting multiple columns, one in descending order, the other in ascending order
sort3 = df.sort_values(by=["names", "city"], ascending=[False, True])
sort3
```

```python
df.sort_values(by=["names", "city"], ascending=[False, True], inplace=True)
df
```
